```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import os
from itertools import combinations
from tqdm import tqdm
os.chdir('D:\Shashi Katteri\Desktop\data sets to work on\Amazon Employee Access\Dataset_10')
```

```python
df=pd.read_csv('Amazon_employee_access.csv')
df.head()
```

```python
cols=list(df.columns)
chisquare_results=pd.DataFrame() 
for i,j in enumerate(combinations(cols,2)):
    ct=pd.crosstab(df[j[0]],df[j[1]])
    chisquare=st.chi2_contingency(ct)
    chisquare_results.loc[i,'Col 1']=j[0]
    chisquare_results.loc[i,'Col 2']=j[1]
    chisquare_results.loc[i,'Chi2_stat']=chisquare[0]
    chisquare_results.loc[i,'p-value']=chisquare[1]
chisquare_results
   
```
